```python
def split_text(df, i):
    text =  df.content[i].replace("\n", " ") 
    data = {"index_id": ["ROW1"],
           "raw_text": text}
    df1 = pd.DataFrame(data).set_index("index_id")
    split_words = [df1.raw_text.str.split()[0][0], "Star", "stars", "ratings", "recent views","Flexible deadlines", "Available languages", "Subtitles"]
    new_df = df1.raw_text.replace("\n", " ").str.extract("(.+)".join(split_words), flags=re.DOTALL)
    new_df1 = new_df.drop(new_df.columns[[0, 2, 3, 5]], axis=1)
    new_df1.columns = ['rating', 'Description','language']
    return new_df1
for i in range(len(df)):
    new_df1 = split_text(df,i)
    print(i, new_df1)
    
```

```python
new_df = pd.DataFrame([])
for i in range(len(df)):
    new_df1 = split_text(df,i)
    new_df = new_df.append(new_df1, ignore_index = True)
df2 = pd.concat([df, new_df], axis=1, join='inner')
```
